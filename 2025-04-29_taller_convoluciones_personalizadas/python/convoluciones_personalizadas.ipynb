# üß™ **Filtro Visual : Convoluciones Personalizadas**

**Computaci√≥n visual**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, VBox # Seccion interactiva
import imageio
from PIL import Image
import os
```

## ‚úÖ Funci√≥n convolucional

```python
def aplicar_convolucion_manual(img, kernel):
    """
    Aplica un filtro definido por un kernel a una imagen en escala de grises.

    Par√°metros:
    - img: imagen de entrada (en escala de grises).
    - kernel: matriz que define el filtro (2D numpy array).

    Retorna:
    - Imagen filtrada por convoluci√≥n.
    """
    h, w = img.shape
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2

    # A√±adir bordes para evitar p√©rdida de informaci√≥n en los extremos
    img_padded = cv2.copyMakeBorder(img, pad_h, pad_h, pad_w, pad_w, cv2.BORDER_REPLICATE)

    # Crear imagen de salida vac√≠a
    output = np.zeros_like(img)

    # Aplicar convoluci√≥n pixel por pixel
    for i in range(h):
        for j in range(w):
            region = img_padded[i:i+kh, j:j+kw]
            output[i, j] = np.clip(np.sum(region * kernel), 0, 255)

    return output
```

## ‚úÖ Cargar imagen en escala de grises

```python
img = cv2.imread('../datos/mariposa.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title("Imagen original")
plt.axis('off')
plt.show()
```

## ‚úÖ Definici√≥n de kernels

```python
# Filtro de enfoque (sharpen)
sharpen_kernel = np.array([
    [0, -1,  0],
    [-1, 5, -1],
    [0, -1,  0]
])

# Filtro de suavizado (blur)
blur_kernel = np.ones((3, 3), dtype=np.float32) / 9.0

# Filtro de detecci√≥n de bordes (Sobel horizontal)
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
```

## ‚úÖ Aplicar filtros manuales

```python
img_sharpen = aplicar_convolucion_manual(img, sharpen_kernel)
img_blur = aplicar_convolucion_manual(img, blur_kernel)
img_edges = aplicar_convolucion_manual(img, sobel_x)
```

## ‚úÖ Aplicar filtros integrados en OpenCV

```python
img_sharpen_cv = cv2.filter2D(img, -1, sharpen_kernel)
img_blur_cv = cv2.filter2D(img, -1, blur_kernel)
img_edges_cv = cv2.filter2D(img, -1, sobel_x)
```

## ‚úÖ Generaci√≥n de imagen comparativa

```python
titulos = ['Sharpen Manual', 'Sharpen OpenCV',
           'Blur Manual', 'Blur OpenCV',
           'Sobel Manual', 'Sobel OpenCV']

imagenes = [img_sharpen, img_sharpen_cv,
            img_blur, img_blur_cv,
            img_edges, img_edges_cv]

plt.figure(figsize=(9, 7))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(imagenes[i], cmap='gray')
    plt.title(titulos[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## ‚úÖ Generar GIF para evidencia de resultados

```python
# Crear carpeta de resultados si no existe
os.makedirs("resultados", exist_ok=True)

# -------------------------------------
# GIF 1: Evoluci√≥n secuencial de im√°genes
# -------------------------------------

# Lista ordenada de im√°genes: original + convoluciones cl√°sicas + convoluciones manuales
secuencia = [
    ("Imagen original", img),
    ("Filtro Sharpen", img_sharpen),
    ("Filtro Sharpen manual", img_sharpen_cv),
    ("Filtro Blur cl√°sico", img_blur),
    ("Filtro Blur manual", img_blur_cv),
    ("Filtro Sobel cl√°sico", img_edges),
    ("Filtro Sobel manual", img_edges_cv)
]

frames = []
for i, (titulo, imge) in enumerate(secuencia):
    fig, ax = plt.subplots()
    ax.imshow(imge, cmap='gray')
    ax.set_title(titulo)
    ax.axis('off')
    
    # Guardar imagen temporal
    temp_path = f"resultados/temp_{i}.png"
    fig.savefig(temp_path, bbox_inches='tight')
    plt.close(fig)

    # Convertir a frame de GIF
    frame = Image.open(temp_path)
    frames.append(frame)

# Guardar GIF final
gif_path1 = "resultados/proceso_convoluciones.gif"
frames[0].save(gif_path1, save_all=True, append_images=frames[1:], duration=1000, loop=0)

# -------------------------------------
# GIF 2: Comparaciones (original vs cl√°sico vs manual)
# -------------------------------------

comparaciones = [
    ("Sharpen", img_sharpen_cv, img_sharpen),
    ("Blur", img_blur_cv, img_blur),
    ("Sobel", img_edges_cv, img_edges)
]

frames_comp = []

for i, (nombre, clasico, manual) in enumerate(comparaciones):
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    titles = ["Original", f"{nombre} Cl√°sico", f"{nombre} Manual"]
    images = [img, clasico, manual]

    for ax, title, imge in zip(axes, titles, images):
        ax.imshow(imge, cmap='gray')
        ax.set_title(title)
        ax.axis('off')

    # Guardar imagen temporal
    temp_path = f"resultados/comp_temp_{i}.png"
    fig.savefig(temp_path, bbox_inches='tight')
    plt.close(fig)

    # Convertir a frame de GIF
    frame = Image.open(temp_path)
    frames_comp.append(frame)

# Guardar segundo GIF
gif_path2 = "resultados/comparaciones_convoluciones.gif"
frames_comp[0].save(gif_path2, save_all=True, append_images=frames_comp[1:], duration=1500, loop=0)
```

## ‚úÖ Secci√≥n interactivo con sliders

```python
# Funci√≥n para aplicar kernel personalizado
# Crear una interfaz interactiva con sliders (`cv2.createTrackbar`) para probar diferentes valores del kernel.

def aplicar_kernel(k00, k01, k02, k10, k11, k12, k20, k21, k22, divisor):
    kernel = np.array([
        [k00, k01, k02],
        [k10, k11, k12],
        [k20, k21, k22]
    ], dtype=np.float32)

    if divisor == 0:
        divisor = 1
    kernel /= divisor

    resultado = cv2.filter2D(img, -1, kernel)

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(resultado, cmap='gray')
    plt.title("Filtrada")
    plt.axis("off")
    plt.show()

# Sliders para cada valor del kernel + divisor
sliders = [
    IntSlider(min=-10, max=10, value=0, description=f'K{i}') for i in range(9)
] + [IntSlider(min=1, max=20, value=1, description='Div')]

# Mostrar interfaz interactiva
interact(
    aplicar_kernel,
    k00=sliders[0], k01=sliders[1], k02=sliders[2],
    k10=sliders[3], k11=sliders[4], k12=sliders[5],
    k20=sliders[6], k21=sliders[7], k22=sliders[8],
    divisor=sliders[9]
);
```
